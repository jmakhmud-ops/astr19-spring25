This notebook defines functions for computing `sin(x)` and `cos(x)`, tabulates their values over a range, and prints the first 10 entries.

## Cell 1: Define the `sin(x)` Function

The function `compute_sin(x)` takes a single float value `x` (in radians) and returns the sine of that value using Python's built-in `math` library.

- **Input:** `x` — a float representing an angle in radians
- **Output:** `math.sin(x)` — the sine of x, a float between -1 and 1

```python
import math

def compute_sin(x):
    """
    Returns the sine of x.
    Parameters: x (float) - angle in radians
    Returns: float - sine of x
    """
    return math.sin(x)

# Quick test
print("sin(0)  =", compute_sin(0))
print("sin(π/2)=", compute_sin(math.pi / 2))
```

## Cell 2: Define the `cos(x)` Function

The function `compute_cos(x)` takes a single float value `x` (in radians) and returns the cosine of that value using Python's built-in `math` library.

- **Input:** `x` — a float representing an angle in radians
- **Output:** `math.cos(x)` — the cosine of x, a float between -1 and 1

```python
def compute_cos(x):
    """
    Returns the cosine of x.
    Parameters: x (float) - angle in radians
    Returns: float - cosine of x
    """
    return math.cos(x)

# Quick test
print("cos(0)  =", compute_cos(0))
print("cos(π/2)=", compute_cos(math.pi / 2))
```

## Cell 3: Tabulate sin(x) and cos(x)

Here we generate **1000 evenly spaced values** of `x` between 0 and 2, and compute both `sin(x)` and `cos(x)` for each value using the functions defined above.

The results are stored in three lists:
- `x_values` — the x values
- `sin_values` — the corresponding sin(x) values
- `cos_values` — the corresponding cos(x) values

```python
# Parameters
start = 0
end = 2
num_entries = 1000

# Generate x values and compute sin/cos
x_values   = []
sin_values = []
cos_values = []

step = (end - start) / (num_entries - 1)

for i in range(num_entries):
    x = start + i * step
    x_values.append(x)
    sin_values.append(compute_sin(x))
    cos_values.append(compute_cos(x))

print(f"Table generated with {len(x_values)} entries from x = {x_values[0]} to x = {x_values[-1]:.6f}")
```

## Cell 4: Print the First 10 Values

Using a `for` loop, we print the first 10 entries of the table in neatly formatted columns showing `x`, `sin(x)`, and `cos(x)`.

```python
# Print header
print(f"{'x':>12}  {'sin(x)':>15}  {'cos(x)':>15}")
print("-" * 46)

# Print first 10 values using a for loop
for i in range(10):
    print(f"{x_values[i]:>12.6f}  {sin_values[i]:>15.10f}  {cos_values[i]:>15.10f}")
```
